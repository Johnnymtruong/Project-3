```python
import pandas as pd
```

Four csv's of pertinent data have been selected. The chosen date range for all data must be on or after 1990, with countries beign the focus of the data visualization. Cleaned csv's are prepped for database insertion.


```python
pd.options.display.max_colwidth = 200
death_rate_df = pd.read_csv("Original_Datasets/death_rate.csv")
death_rate_df
```

```python
death_rate_copy = death_rate_df.copy()
death_rate_copy
```

```python
death_rate_copy.dtypes
```

```python
death_rate_copy.info
```

```python
death_rate_index = death_rate_copy[(death_rate_copy['Year'] <1990)|(death_rate_copy['Country_name'].str.contains('WB'))|
                                   (death_rate_copy['Country_name'].str.contains('World'))|
                                   (death_rate_copy['Country_name'].str.contains('Income'))|
                                   (death_rate_copy['Country_name'].str.contains('income'))|
                                  (death_rate_copy['Country_name'].str.contains('Region'))|
                                   (death_rate_copy['Country_name'].str.contains('Countries'))|
                                  (death_rate_copy['Country_name'].str.contains('countries'))].index
death_rate_copy.drop(death_rate_index, inplace = True)
death_rate_copy
```

```python
death_country_index = death_rate_copy[(death_rate_copy['Country_name'] == 'World')].index
death_rate_copy.drop(death_country_index, inplace = True)
death_rate_copy
```

```python
death_rate_copy['Country_name'].value_counts()
```

```python
death_rate_copy['Year'].value_counts()
```

```python
death_rate_copy.reset_index(drop=True)
```

```python
death_rate_copy.to_csv('Cleaned_Datasets/death_rate_clean', sep=',', index=False, encoding='utf-8')
```

```python
pd.options.display.max_colwidth = 200
gdp_df = pd.read_csv("Original_Datasets/gdp_per_capita.csv")
gdp_df
```

```python
gdp_clean.reset_index(drop=True)
```

```python
pd.options.display.max_colwidth = 200
deaths_df = pd.read_csv("Original_Datasets/Number_of_Deaths_by_Risk_Factors.csv")
deaths_df
```

```python
deaths_df.columns
```

```python
deaths_copy = deaths_df.copy()
```

```python
deaths_copy
```

```python
death_renamed = deaths_copy.rename(columns = {"Entity":"Country_name", " Smoking":"Smoking", 
                                              "Household_air_pollution_from_solid_fuels": "Household_air_pollution",
                                              "No_access_to_handwashing_facility":"No_handwash_access", 
                                              "Drug_use_Sex_Both_Age_All_Ages_Number":"Drug_Use",
                                             "Low_bone_mineral_density":"Low_bone_density"})
death_renamed
```

```python
death_index = death_renamed[(death_renamed['Year'] <1990) | (death_renamed['Country_name'].str.contains('World'))|
                            (death_renamed['Country_name'].str.contains('Region'))
                           |(death_renamed['Country_name'].str.contains('WHO'))|
                            (death_renamed['Country_name'].str.contains('income'))|
                           (death_renamed['Country_name'].str.contains('countries'))].index
death_renamed.drop(death_index, inplace = True)
death_renamed
```

```python
death_renamed.dtypes
```

```python
death_renamed.info()
```

```python
death_renamed.reset_index(drop=True)
```

```python
death_renamed.to_csv('Cleaned_Datasets/death_risks_clean', sep=',', index=False, encoding='utf-8')
```
