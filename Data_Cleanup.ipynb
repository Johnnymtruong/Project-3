```python
import pandas as pd
from json import loads, dumps
import numpy as np
```

Four csv's of pertinent data have been selected. The chosen date range for all data must be on or after 1990, with countries beign the focus of the data visualization. Cleaned csv's are prepped for database insertion.


```python
pd.options.display.max_colwidth = 200
death_rate_df = pd.read_csv("Original_Datasets/death_rate.csv")
death_rate_df
```

```python
death_rate_copy = death_rate_df.copy()
death_rate_copy
```

```python
death_rate_copy.dtypes
```

```python
death_rate_copy.info
```

```python
death_rate_index = death_rate_copy[(death_rate_copy['Year'] <1990)|(death_rate_copy['Country_name'].str.contains('WB'))|
                                   (death_rate_copy['Country_name'].str.contains('World'))|
                                   (death_rate_copy['Country_name'].str.contains('Income'))|
                                   (death_rate_copy['Country_name'].str.contains('income'))|
                                  (death_rate_copy['Country_name'].str.contains('Region'))|
                                   (death_rate_copy['Country_name'].str.contains('Countries'))|
                                  (death_rate_copy['Country_name'].str.contains('countries'))].index
death_rate_copy.drop(death_rate_index, inplace = True)
death_rate_copy
```

```python
death_country_index = death_rate_copy[(death_rate_copy['Country_name'] == 'World')].index
death_rate_copy.drop(death_country_index, inplace = True)
death_rate_copy
```

```python
death_rate_copy['Country_name'].value_counts()
```

```python
death_rate_copy['Year'].value_counts()
```

```python
death_rate_updated = death_rate_copy.reset_index(drop=True)
death_rate_updated
```

```python
death_rate_updated['ID'] = ''
death_rate_updated
```

```python
death_order = death_rate_updated.iloc[:,[3,0,1,2]]
death_order
```

```python
death_order['ID'] = range(len(death_order))
death_order
```

```python
death_order.to_json('Cleaned_Datasets/json/death_rate_clean.json', orient= 'index')
```

```python
death_order.to_csv('Cleaned_Datasets/death_rate_clean.csv', sep=',', index=False, encoding='utf-8')
```

```python
result = death_order.to_json(orient= 'index')
parsed = loads(result)
dumps(parsed)
```
